# データ探索、理解のためのコード

```python
# サンプル
# 必要なライブラリをインポート
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
# データの読み込み
df = pd.read_csv('/workspaces/boatrace_show_prediction/data/raw/merged_data.csv')

# データの基本情報を表示
print(df.info())
print(df.describe())
```

```python
tamagawa_count = df[df["レース場"]=="多摩川"].shape[0]
print(tamagawa_count)
```

```python
# 欠損値の確認
null_count = df["天気"].isnull().sum()
print(null_count)

## 全体
# print(df.isnull().sum())
```

```python
# 天気に欠損値が含まれるレースをフィルタリング
missing_weather_races = df[df['天気'].isnull()]

# 欠損値が含まれるレースを表示
print(missing_weather_races.head())  # 最初の数行を表示
```

```python
data_types = df.dtypes

# データタイプでグループ分け
columns_by_type = data_types.groupby(data_types).apply(lambda x: list(x.index))
columns_by_type_dict = columns_by_type.to_dict()
columns_by_type_dict
```

```python
# カテゴリ変数の分布を確認
sns.countplot(x='category_feature', data=df)
plt.show()
```

```python
# 数値変数のヒストグラムを表示
df['numeric_feature'].hist(bins=30)
plt.xlabel('numeric_feature')
plt.ylabel('Frequency')
plt.show()
```

```python
# 相関行列の表示
corr_matrix = df.corr()
sns.heatmap(corr_matrix, annot=True)
plt.show()
```
