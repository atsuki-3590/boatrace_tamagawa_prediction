データ探索、理解のためのコード

```python
# サンプル
# 必要なライブラリをインポート
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# データの読み込み
df = pd.read_csv('data/processed/cleaned_data.csv')

# データの基本情報を表示
print(df.info())
print(df.describe())
```

```python
# 欠損値の確認
print(df.isnull().sum())
```

```python
# カテゴリ変数の分布を確認
sns.countplot(x='category_feature', data=df)
plt.show()
```

```python
# 数値変数のヒストグラムを表示
df['numeric_feature'].hist(bins=30)
plt.xlabel('numeric_feature')
plt.ylabel('Frequency')
plt.show()
```

```python
# 相関行列の表示
corr_matrix = df.corr()
sns.heatmap(corr_matrix, annot=True)
plt.show()
```
